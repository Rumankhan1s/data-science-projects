```python
import pandas as pd

data = {
    "Order_ID": [101,102,103,104,105,106,107,108],
    "Product": ["Laptop","Phone","Tablet","Laptop","Phone","Tablet","Laptop","Phone"],
    "Price": [50000,20000,None,52000,21000,16000,51000,22000],
    "Quantity": [3,5,2,None,6,3,2,4],
    "City": ["Delhi","Mumbai","Delhi","Jaipur","Mumbai","Delhi","Mumbai","Jaipur"]
}

df = pd.DataFrame(data)
df
```

```python
df["Price"] = df["Price"].fillna(df["Price"].mean())
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
```

```python
df["Revenue"] = df["Price"] * df["Quantity"]
df
```

```python
product_sales = df.groupby("Product")["Revenue"].sum()
product_sales
```

```python
city_sales = df.groupby("City")["Revenue"].sum()
city_sales
```

```python
df["Revenue"].sum()
```

```python
import matplotlib.pyplot as plt

product_sales.plot(kind="bar")
plt.title("Revenue by Product")
plt.ylabel("Revenue")
plt.show()
```

```python
import matplotlib.pyplot as plt

city_sales.plot(kind="bar")
plt.title("Revenue by Product")
plt.ylabel("Revenue")
plt.show()
```
